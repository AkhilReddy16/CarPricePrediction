## common import statements

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot
```

## Get data from website and assign appropriate headers.

```python
# Importing the Data from the website,
other_path = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
data = pd.read_csv(other_path, names=headers)
data.head(10)
```

## checking data types to get an idea on the data

```python
data.dtypes #to know type of each coloumn.
```

## Remove ? and replace them with NAN values.

```python
# Replace "?" to NaN,
data.replace("?", np.nan, inplace = True)
data.head(5)
```

## Create a table with true or false to see the null values and count null values in each coloumn.

```python
is_data_null = data.isnull() #this will create a new table saying wheather the coresponding value is null or not.
is_data_null.head(5)
for col in is_data_null.columns.values.tolist():
    print(is_data_null[col].value_counts())
    print("")  
```

Replace by mean:

"normalized-losses": 41 missing data, replace them with mean "stroke": 4 missing data, replace them with mean "bore": 4 missing data, replace them with mean "horsepower": 2 missing data, replace them with mean "peak-rpm": 2 missing data, replace them with mean Replace by frequency

"num-of-doors": 2 missing data, replace them with "four". Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur Drop the whole row

"price": 4 missing data, simply delete the whole row Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us.

```python
mean_fill_col = ["normalized-losses","bore","stroke","horsepower","peak-rpm"]
for col in mean_fill_col:
    mean = data[col].astype("float").mean(axis=0)
    data[col].replace(np.nan, mean, inplace=True)
data["num-of-doors"].replace(np.nan, "four", inplace=True)
data.dropna(subset=["price"], axis=0, inplace=True) #we don't need rows that don't have price which is our feature.
data.reset_index(drop=True, inplace=True)
```

## convert numerical data in object form to int or flot

```python
#Converting the datatypes,
data[["bore", "stroke","price","peak-rpm"]] = data[["bore", "stroke","price","peak-rpm"]].astype("float")
data[["normalized-losses","horsepower"]] = data[["normalized-losses","horsepower"]].astype("int")
data.dtypes
```

## Scale length, width, & height to have variance as 1.

```python
#scaling the variable so the variable average is 0, scaling the variable so the variance is 1, 
#or scaling variable so the variable values range from 0 to 1 this is to normalize the data.
data["length"] = data["length"]/data["length"].max()
data["width"] = data["width"]/data["width"].max()
data['height'] = data['height']/data['height'].max() 
data[["length","width","height"]].head()
```

## Convert fuel-type coloumn into dummies to get data in numerical format.

```python
dummy = pd.get_dummies(data["fuel-type"])
dummy.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
data = pd.concat([data, dummy], axis=1)
data.drop("fuel-type", axis = 1, inplace=True)
data.head(5)
```

```python
dummy = pd.get_dummies(data["aspiration"])
dummy.rename(columns={'std':"aspiration-std",'turbo':"aspiration-turbo"}, inplace=True)
data = pd.concat([data, dummy], axis=1)
data.drop("aspiration", axis = 1, inplace=True)
data.head(5)
```

## Drop all catagorical data to only have numerical data. 

```python
data = data.select_dtypes(exclude=['object'])
```

## rearrange coloumn position in the data.

```python
X=data.drop('price', axis=1)
y=data[['price']]
data = pd.concat([X, y], axis=1)
data.head(5)
```

## export processed data to csv

```python
data.to_csv('processed_data.csv')
```
